```python
quan=[]
qual=[]
for columnname in dataset.columns:
    if dataset[columnname].dtype=='O':
        qual.append(columnname)
    else:
        quan.append(columnname)

    
```

## Imputation of Numerical data by replacing with Mean, Mode and Median

```python
table1=pd.DataFrame(data=dataset[quan],columns=quan)
table1
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=pd.NA,strategy='mean')
```

```python
quan_dataset=imputer.fit_transform(table1)
```

```python
quan_dataset
```

```python
quan_table=pd.DataFrame(data=quan_dataset,columns=quan)
```

```python
quan_table
```

```python
quan_table.isnull().sum()
```

## Imputation of Categorical data with Mode

```python
table2=pd.DataFrame(data=dataset[qual],columns=qual)
table2
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='most_frequent')
```

```python
qual_dataset=imputer.fit_transform(dataset[qual])
```

```python
qual_dataset
```

```python
qual_table=pd.DataFrame(data=qual_dataset, columns=qual)
qual_table
```

```python
qual_table.isnull().sum()
```
